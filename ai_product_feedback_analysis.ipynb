# AI Product Analytics – User Feedback & Insight System

This notebook analyzes 2,000+ user feedback records to identify high-severity issues,
prioritize product improvements, and generate data-driven insights.

```python
import pandas as pd

df = pd.read_csv('user_feedback_dataset.csv')
df.head()
```

## Data Cleaning & Feature Engineering

```python
severity_map = {'Low': 1, 'Medium': 2, 'High': 3}
df['severity_score'] = df['severity'].map(severity_map)

df.head()
```

## Priority Analysis

```python
priority_df = (
    df.groupby('feedback_type')
      .agg(issue_count=('severity_score', 'count'), avg_severity=('severity_score', 'mean'))
      .assign(priority_score=lambda x: x.issue_count * x.avg_severity)
      .sort_values('priority_score', ascending=False)
)

priority_df.head()
```
